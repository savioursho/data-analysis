# 分布のプロット

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

## ガンマ分布

```python
x = np.linspace(0, 20, 200)
fig, ax = plt.subplots(1, 2, figsize=(10,4))
list_nu = [1, 2, 3, 5, 10, 15]
list_alpha = [0.5, 1, 1.5, 2, 2.5, 3]


for nu in list_nu:
    alpha = 1
    array_gamma = stats.gamma.pdf(x, nu, scale=alpha)
    ax[0].plot(x, array_gamma, label=f"nu={nu}, alpha={alpha}")
for alpha in list_alpha:
    nu = 5
    array_gamma = stats.gamma.pdf(x, nu, scale=alpha)
    ax[1].plot(x, array_gamma, label=f"nu={nu}, alpha={alpha}")
fig.suptitle("Probability Density Function of Gamma Distribution")
ax[0].set(title=r"varying shape parameter $\nu$")
ax[0].legend()
ax[1].set(title=r"varying scale parameter $\alpha$")
ax[1].legend()
fig.tight_layout();
```
